```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

X, y= make_blobs(n_samples=100,  centers=1, cluster_std=0.4, random_state=0)#100 obseravtion dans les données
print("X=\n",X,"\n y=\n",y)

X[-1,:]=np.array([3,7]) #une anomalie est définie aux coordonnées (x,y)= (3,7) {dans la dernière ligne de X}
plt.scatter(X[:,0],X[:,1])
#Mise en oeuvre de l'algo de Isolation Forest
```

```python
from sklearn.ensemble import IsolationForest
import numpy as np
model = IsolationForest(contamination=0.1,n_estimators=100) #on veut filter 1% de donnée
model.fit(X)
plt.scatter(X[:,0],X[:,1], c=model.predict(X)) #predire l'outlier data,
#l'anomalie,si elle est correctement prédite, appariatera d'une couleur différence des autres,
```

```python
from sklearn.ensemble import IsolationForest
clf = IsolationForest(max_samples=100, random_state=4, contamination=.1)
#identify outliers:
y_pred_train = clf.fit_predict(X)
#Remove outliers where 1 represent inliers and -1 represent outliers:
X_train_cleaned = X[np.where(y_pred_train == 1, True, False)]
plt.scatter(X_train_cleaned[:,0],X_train_cleaned[:,1])
```
