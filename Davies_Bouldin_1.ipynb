```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import davies_bouldin_score
from scipy. cluster.hierarchy import dendrogram, linkage

#load wine data set
diab = load_diabetes()
data=diab.data

#Perform agglomerative Hierarcical Clustering
agg_clustering = AgglomerativeClustering(n_clusters=3)
agg_clustering.fit(data)

#Get the cluster labels
labels = agg_clustering.labels_

#Calculate Davies-Bouldin Index
db_index=davies_bouldin_score(data, labels)
print("Diabeties Data:\n", data)
print(f"Davies-Bouldin Index:{db_index}")

#Create  a linkage matrix for dendrogram
linkage_matrix=linkage(data,method='ward')
#Plot the dendrogram
plt.figure(figsize=(12,6)) #create a new figure
dendrogram(linkage_matrix, orientation="top",labels=labels,distance_sort='descending')
plt.title('Dendrogram Asma')
plt.show()


"""dendrogram: Plot the hierarchical clustering as a dendrogram.

The dendrogram illustrates how each cluster is
composed by drawing a U-shaped link between a non-singleton
cluster and its children"""
```
